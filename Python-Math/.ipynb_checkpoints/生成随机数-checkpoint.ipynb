```python
# -*- coding: utf-8 -*-
__author__ = 'lijianming'
import random, math

# 正态分布
print [random.normalvariate(0, 1) for i in range(10000)]
print [random.gauss(1, 0.5) for i in range(10000)]

# 对数正态分布
print [random.lognormvariate(0, 1) for i in range(10000)]
print [random.lognormvariate(0, 0.5) for i in range(10000)]

# 冯.米塞斯分布（循环正态分布）
print [random.vonmisesvariate(math.pi, 0.5) for i in range(10000)]
print [random.vonmisesvariate(math.pi, 2) for i in range(10000)]

# 三角形分布
print [random.triangular(0, 1, 0.3) for i in range(10000)]
print [random.triangular(0.5, 2.5, 2) for i in range(10000)]

# 指数分布
print [random.expovariate(1.5) for i in range(10000)]
print [random.expovariate(0.5) for i in range(10000)]

# 伽马分布
print [random.gammavariate(1, 0.5) for i in range(10000)]
print [random.gammavariate(5, 1) for i in range(10000)]

# 贝塔分布
print [random.betavariate(2, 2) for i in range(10000)]
print [random.betavariate(0.5, 0.5) for i in range(10000)]

# 威布尔分布
print [random.weibullvariate(1, 5) for i in range(10000)]
print [random.weibullvariate(1, 1.5) for i in range(10000)]
```
