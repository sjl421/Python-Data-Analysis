# 一、Series是一种类似于一维数组的对象，他有一组数据（各种Nupython数据类型），以及一组与之相关的数据标签（索引）

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

print '生成最简单的Series'
obj = Series([4, 7, -5, 3])
print '显示Series'
print obj
print '获取索引'
print obj.index
print '获取数据'
print obj.values

print '生成Series并自定义索引'
obj2 = Series([4, 7, -5, 3], index=['d', 'b', 'a', 'c'])
print '显示Series'
print obj2
print'通过索引查找值'
print obj2['a']
print 'Series赋值'
obj2['d'] = 6
print'通过索引查找多个值'
obj2[['c', 'a', 'd']]

print '支持Numpy运算'
print '大于0的值'
print obj2[obj2 > 0]
print '所有值乘以2'
print obj2 * 2
print '指数运算'
print np.exp(obj2)
print '布尔运算'
print 'b' in obj2, 'e' in obj2

print '通过字典创建Series'
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = Series(sdata)

print '使用原Series，通过传入key生成新Series'
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj4 = Series(sdata, index=states)
print obj4

print '判断有没有空'
print pd.isnull(obj4), obj4.isnull()
print '判断有没有非空'
print pd.notnull(obj4)

print '两个Series相同Key，value相加'
print obj3 + obj4

print 'Series对象本身及索引都有name属性'
obj4.name = 'population'
obj4.index.name = 'state'
print obj4

print '索引的值可以修改'
obj.index = ['Bob', 'Steve', 'Jeff', 'Ryan']
print obj
```
